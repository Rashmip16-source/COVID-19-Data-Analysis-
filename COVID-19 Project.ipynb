### Data Visualization

### For India

```python
grp_data.loc['India'].plot(color = 'r')
plt.title('COVID Casas Trend Over a Time')
plt.legend()
plt.grid()
plt.show()
```

```python
grp_data.loc['India'][0:11].plot(color = 'r', marker = 'o')
plt.title('COVID Casas Trend Over a Time')
plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,6))
grp_data.loc['India'][15:50].plot(color = 'r', marker = 'o')
plt.title('COVID Casas Trend Over a Time')
plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,6))
grp_data.loc['India'][40:70].plot(color = 'm', marker = 'o')
plt.title('COVID Casas Trend Over a Time')
plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.xticks(np.arange(0,31))
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
grp_data.loc['India'][70:].plot(color = 'm', marker = 'o')
plt.title('COVID Casas Trend Over a Time')
plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.xticks(np.arange(0,31))
plt.yticks(np.arange(0,36000, 1000))
plt.legend()
plt.grid()
plt.show()
```

### For China
